```python
#Import my favourite librarys :)
import numpy as np
from scipy.optimize import curve_fit as cf
import matplotlib.pyplot as plt
#Import the data from CSV
data20=np.genfromtxt("nitrogen_20ml.csv",skip_header=2,delimiter=",")
data40=np.genfromtxt("nitrogen_40ml.csv",skip_header=2,delimiter=",")
data60=np.genfromtxt("nitrogen_60ml.csv",skip_header=2,delimiter=",")
data80=np.genfromtxt("nitrogen_80ml.csv",skip_header=2,delimiter=",")
data100=np.genfromtxt("nitrogen_100ml.csv",skip_header=2,delimiter=",")
```

```python
#Plot all the data so we know whats going on
fig, (ax1, ax2, ax3, ax4, ax5) =plt.subplots(5, figsize=(15,15))
ax1.plot(data20[:,0],data20[:,1], "blue", label = "emf for V_eq = 20ml")
ax1.axhline(0, color = "black", linewidth = 0.5)
ax1.set_ylabel("emf (V)")
ax1.legend(loc = "upper right")
ax1.set_xlabel("Time / s")
ax1.set_xlim(xmin=0)

ax2.plot(data40[:,0],data40[:,1], "orange", label = "emf for V_eq = 40ml")
ax2.axhline(0, color = "black", linewidth = 0.5)
ax2.set_ylabel("emf (V)")
ax2.legend(loc = "upper right")
ax2.set_xlabel("Time / s")
ax2.set_xlim(xmin=0)

ax3.plot(data60[:,0],data60[:,1], "red", label = "emf for V_eq = 60ml")
ax3.axhline(0, color = "black", linewidth = 0.5)
ax3.set_ylabel("emf (V)")
ax3.legend(loc = "upper right")
ax3.set_xlabel("Time / s")
ax3.set_xlim(xmin=0)

ax4.plot(data80[:,0],data80[:,1], "green", label = "emf for V_eq = 80ml")
ax4.axhline(0, color = "black", linewidth = 0.5)
ax4.set_ylabel("emf (V)")
ax4.legend(loc = "upper right")
ax4.set_xlabel("Time / s")
ax4.set_xlim(xmin=0)

ax5.plot(data100[:,0],data100[:,1], "black", label = "emf for V_eq = 100ml")
ax5.axhline(0, color = "black", linewidth = 0.5)
ax5.set_ylabel("emf (V)")
ax5.legend(loc = "upper right")
ax5.set_xlabel("Time / s")
ax5.set_xlim(xmin=0)
```

```python
def velocityfunction (t,c,F,w):
    return c*F*w*np.sin(w*t)*np.exp(-c*t)
```

```python
popt, pcov = cf(velocityfunction, data20[:,0],data20[:,1])
print(popt)
```

```python
popt, pcov = cf(velocityfunction, data40[:,0],data40[:,1])
print(popt)
```
