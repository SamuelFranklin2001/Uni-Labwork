```python
#I aim to plot omega against the reciprocal of the volume so i will create a function that takes in an array from above
def w_v(data, volume):
    uncertainty=[]
    for i in range(len(data[1])):
        uncertainty.append(20*10**-6)
    
    popt,pcov=cf(fit, data[0],data[1],maxfev = 30000,sigma=uncertainty)
    
    if(pcov[0,0]<300 and pcov[0,0]>-300):
        omega_squared=(popt[0])**2
    
        omega_err=((pcov[0,0]**0.5)/popt[0])*2*omega_squared
    
        reciprocal_volume=((volume*10**-6)+(6.3*10**-6))**-1
        v_err=(10**-6/volume)*-1*reciprocal_volume
        return reciprocal_volume, v_err, omega_squared, omega_err
    else:
        return False
```

```python
#Now get the omega squared and the volume reciprocal as data points and put them in an array
#         V    verr  w Sq    werr
N2_v_w=[  [] , [],   []    ,  []  ]
He_v_w=[  [] , [],   []    ,  []  ]
CO2_v_w=[ [] , [],   []   ,   []  ]
Air_v_w=[ [] , [],   []   ,   []  ]

    
#make a general procedure that fills up the created arrays for the gasses with the angular frequency squared and reciprocal volume

def w_v_fill(input_data_array, output_data_array,volume):
    for i in range (len(input_data_array)):
        if(w_v(input_data_array[i],volume)!=False):
            recip_vol, vol_err, w_sq, w_err= w_v(input_data_array[i],volume)
            output_data_array[0].append(recip_vol)
            output_data_array[1].append(vol_err)
            output_data_array[2].append(w_sq)
            output_data_array[3].append(w_err)
            
            
#Fill my arrays

## Nitrogen
w_v_fill(N2_10_clean,N2_v_w,10)
w_v_fill(N2_20_clean,N2_v_w,20)
w_v_fill(N2_30_clean,N2_v_w,30)
w_v_fill(N2_40_clean,N2_v_w,40)
w_v_fill(N2_50_clean,N2_v_w,50)
w_v_fill(N2_60_clean,N2_v_w,60)
w_v_fill(N2_70_clean,N2_v_w,70)
w_v_fill(N2_80_clean,N2_v_w,80)
w_v_fill(N2_90_clean,N2_v_w,90)
w_v_fill(N2_100_clean,N2_v_w,100)

##Carbon Dioxide
w_v_fill(CO2_10_clean,CO2_v_w,10)
w_v_fill(CO2_20_clean,CO2_v_w,20)
w_v_fill(CO2_30_clean,CO2_v_w,30)
w_v_fill(CO2_40_clean,CO2_v_w,40)
w_v_fill(CO2_50_clean,CO2_v_w,50)
w_v_fill(CO2_60_clean,CO2_v_w,60)
w_v_fill(CO2_70_clean,CO2_v_w,70)
w_v_fill(CO2_80_clean,CO2_v_w,80)
w_v_fill(CO2_90_clean,CO2_v_w,90)
w_v_fill(CO2_100_clean,CO2_v_w,100)

##Helium
w_v_fill(He_10_clean,He_v_w,10)
w_v_fill(He_20_clean,He_v_w,20)
w_v_fill(He_30_clean,He_v_w,30)
w_v_fill(He_40_clean,He_v_w,40)
w_v_fill(He_50_clean,He_v_w,50)
w_v_fill(He_60_clean,He_v_w,60)
w_v_fill(He_70_clean,He_v_w,70)
w_v_fill(He_80_clean,He_v_w,80)
w_v_fill(He_90_clean,He_v_w,90)
w_v_fill(He_100_clean,He_v_w,100)

##Air
w_v_fill(Air_10_clean,Air_v_w,10)
w_v_fill(Air_20_clean,Air_v_w,20)
w_v_fill(Air_30_clean,Air_v_w,30)
w_v_fill(Air_40_clean,Air_v_w,40)
w_v_fill(Air_50_clean,Air_v_w,50)
w_v_fill(Air_60_clean,Air_v_w,60)
w_v_fill(Air_70_clean,Air_v_w,70)
w_v_fill(Air_80_clean,Air_v_w,80)
w_v_fill(Air_90_clean,Air_v_w,90)
w_v_fill(Air_100_clean,Air_v_w,100)
```

```python
#These values seem to deviate from the expected values, I belive this is because
#The assumption the the decay constant<<The angular frequency. breaks down at these low volumes. I will show this by plotting
#the (decay constant/the angular frequency) against the volume

def w_c(data, volume):
    
    
    popt,pcov=cf(fit, data[0],data[1],maxfev = 30000)
    
    if(pcov[0,0]<300 and pcov[0,0]>-300):
        omega=popt[0]
        omega_err=pcov[0,0]*0.5
        
        decay=popt[2]
        decay_err=pcov[2,2]
        
        
        c_over_w=np.sqrt((popt[2]/popt[0])**2)
        
        c_w_err=((omega_err/omega)+(decay_err/decay))*c_over_w
        
        return c_over_w, c_w_err, volume*(10**-6), (10**-6)
    else:
        return False
    
N2_c_w=[  [] , [],   []    ,  []  ]
He_c_w=[  [] , [],   []    ,  []  ]
CO2_c_w=[ [] , [],   []   ,   []  ]
Air_c_w=[ [] , [],   []   ,   []  ]    

def w_c_fill(input_data_array, output_data_array,volume):
    for i in range (len(input_data_array)):
        if(w_c(input_data_array[i],volume)!=False):
            c_over_w, c_w_err, vol, vol_err= w_c(input_data_array[i],volume)
            output_data_array[0].append(c_over_w)
            output_data_array[1].append(c_w_err)
            output_data_array[2].append(vol)
            output_data_array[3].append(vol_err)

            
## Nitrogen
w_c_fill(N2_10_clean,N2_c_w,10)
w_c_fill(N2_20_clean,N2_c_w,20)
w_c_fill(N2_30_clean,N2_c_w,30)
w_c_fill(N2_40_clean,N2_c_w,40)
w_c_fill(N2_50_clean,N2_c_w,50)
w_c_fill(N2_60_clean,N2_c_w,60)
w_c_fill(N2_70_clean,N2_c_w,70)
w_c_fill(N2_80_clean,N2_c_w,80)
w_c_fill(N2_90_clean,N2_c_w,90)
w_c_fill(N2_100_clean,N2_c_w,100)

##Carbon Dioxide
w_c_fill(CO2_10_clean,CO2_c_w,10)
w_c_fill(CO2_20_clean,CO2_c_w,20)
w_c_fill(CO2_30_clean,CO2_c_w,30)
w_c_fill(CO2_40_clean,CO2_c_w,40)
w_c_fill(CO2_50_clean,CO2_c_w,50)
w_c_fill(CO2_60_clean,CO2_c_w,60)
w_c_fill(CO2_70_clean,CO2_c_w,70)
w_c_fill(CO2_80_clean,CO2_c_w,80)
w_c_fill(CO2_90_clean,CO2_c_w,90)
w_c_fill(CO2_100_clean,CO2_c_w,100)

##Helium
w_c_fill(He_10_clean,He_c_w,10)
w_c_fill(He_20_clean,He_c_w,20)
w_c_fill(He_30_clean,He_c_w,30)
w_c_fill(He_40_clean,He_c_w,40)
w_c_fill(He_50_clean,He_c_w,50)
w_c_fill(He_60_clean,He_c_w,60)
w_c_fill(He_70_clean,He_c_w,70)
w_c_fill(He_80_clean,He_c_w,80)
w_c_fill(He_90_clean,He_c_w,90)
w_c_fill(He_100_clean,He_c_w,100)

##Air
w_c_fill(Air_10_clean,Air_c_w,10)
w_c_fill(Air_20_clean,Air_c_w,20)
w_c_fill(Air_30_clean,Air_c_w,30)
w_c_fill(Air_40_clean,Air_c_w,40)
w_c_fill(Air_50_clean,Air_c_w,50)
w_c_fill(Air_60_clean,Air_c_w,60)
w_c_fill(Air_70_clean,Air_c_w,70)
w_c_fill(Air_80_clean,Air_c_w,80)
w_c_fill(Air_90_clean,Air_c_w,90)
w_c_fill(Air_100_clean,Air_c_w,100)
            
```
